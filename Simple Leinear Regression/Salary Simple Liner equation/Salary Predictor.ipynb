```python
warnings.filterwarnings("ignore")
```

```python
sns.pairplot(sal_data)
```

```python
sns.lmplot(x="YearsExperience", y="Salary", data=sal_data)
```

```python
import statsmodels.formula.api as smf
```

```python
mod_1 = smf.ols(formula="Salary~YearsExperience", data=sal_data).fit()
```

```python
mod_1.params
```

```python
print("BIC of Model                 :" ,mod_1.bic)
print("R Squared of Model           :" ,mod_1.rsquared)
print("Adjusted R Squared of Model  :" ,mod_1.rsquared_adj)
print("Adjusted R Squared of Model  :" , mod_1.aic)
```

```python
import pickle
```

```python
pickle.dump(mod_1 ,open("Salary Predictor.pkl","wb"))
```
