```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
import numpy as np
import warnings 

from sklearn.preprocessing import LabelEncoder

warnings.filterwarnings("ignore")
ln = LabelEncoder()
nb_gau = GaussianNB()
nb_mul = MultinomialNB()
```

```python
sal_data = pd.read_csv("SalaryData_Train.csv")
```

```python
sal_data.isna().sum()
```

```python
sal_data.info()
```

```python
sal_data.head(2)
```

```python
sal_data.workclass = ln.fit_transform(sal_data.workclass)
sal_data.education = ln.fit_transform(sal_data.education)
sal_data.maritalstatus = ln.fit_transform(sal_data.maritalstatus)
sal_data.occupation = ln.fit_transform(sal_data.occupation)
sal_data.relationship = ln.fit_transform(sal_data.relationship)
sal_data.race = ln.fit_transform(sal_data.race)
sal_data.sex = ln.fit_transform(sal_data.sex)
sal_data.native = ln.fit_transform(sal_data.native)
sal_data.Salary = ln.fit_transform(sal_data.Salary)
```

```python
sal_data.head(2)
```

```python
x = sal_data.drop(labels="Salary",axis=1)
y = sal_data[["Salary"]]
```

### Gaussia NB

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=12,test_size=0.20,stratify=y)
y_train_pred = nb_gau.fit(x_train,y_train).predict(x_train)
print("===========Train Data================")
print("Accurecy Score    ",accuracy_score(y_train,y_train_pred))

print("")
print("===========Test Data================")
y_test_pred = nb_gau.fit(x_train,y_train).predict(x_test)
print("Accurecy Score",accuracy_score(y_test,y_test_pred))
```

##  For MultinomialNB

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=12,test_size=0.20,stratify=y)
y_train_pred = nb_mul.fit(x_train,y_train).predict(x_train)
print("===========Train Data================")
print("Accurecy Score    ",accuracy_score(y_train,y_train_pred))

print("")
print("===========Test Data================")
y_test_pred = nb_gau.fit(x_train,y_train).predict(x_test)
print("Accurecy Score",accuracy_score(y_test,y_test_pred))
```
