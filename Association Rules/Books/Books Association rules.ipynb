### Model Building

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
items = apriori(df = data,min_support=0.020,use_colnames=True)
```

```python
items
```

```python
best_assoc = association_rules(df = items,min_threshold=0.91)
best_assoc
```
