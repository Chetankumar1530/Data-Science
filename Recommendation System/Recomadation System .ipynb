```python
pvt_table = pd.pivot_table(data = data,values = "Book.Rating", index="User.ID", columns="Book.Title",aggfunc="sum").fillna(0)
```

```python
pvt_table.values
```

```python
cosiv_table = 1- pairwise_distances(pvt_table.values,metric="cosine")
cosiv_table
```

```python
cos_table = pd.DataFrame(data = cosiv_table)
cos_table.index = data["User.ID"].unique()
cos_table.columns = data["User.ID"].unique()


cos_table.round(2)
```

```python
top_50_user = cos_table.iloc[0:500,0:500]
top_50_user.idxmax()
```
