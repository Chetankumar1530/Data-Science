```python
import pandas as pd
import scipy.stats as stats
import numpy as np
```

## Cutlet Data

```python
cut_data=pd.read_csv('Cutlets.csv')
cut_data.head()
```

```python
from scipy.stats import ttest_ind
```

```python
statistic,pvalue=ttest_ind(cut_data['Unit A'],cut_data['Unit B'])
print('statistics is {}\n pvalue is {}'.format(statistic,pvalue))
if pvalue<0.05:
    print('\nSignificant difference in diameter of the cutlet (Ha)')
else:
    print('\nNo Significant change in diameter of the cutlet (Ho)')
```

###  For Lab Data

```python
lab_data=pd.read_csv('LabTAT.csv')
lab_data
```

```python
statistic,pvalue=stats.f_oneway(lab_data['Laboratory 1'],lab_data['Laboratory 2'],lab_data['Laboratory 3'],lab_data['Laboratory 4'])

print('statistics is {}\npvalue is {}'.format(statistic,pvalue))
if pvalue<0.05:
    print('There is significant difference in average TAT of reports between the laboratories (Ha)')
else:
    print('There is no significant change in verage TAT of reports between the laboratories (Ho)')
```
