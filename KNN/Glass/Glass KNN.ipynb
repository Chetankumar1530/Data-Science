```python
import pandas as pd
import warnings

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("glass.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.head(1)
```

```python
x= data.drop(labels="Type",axis = 1)
y = data[["Type"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=12,stratify=y)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
y_train_pred = model.predict(x_train)
print("Accurcy Score"  , accuracy_score(y_train,y_train_pred))
```

```python
for i in range (1,41):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    y_train_pred = model.predict(x_train)
    print(" For ",i," Negn Accurcy Score"  , accuracy_score(y_train,y_train_pred))
```

```python
for i in range (1,41):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    y_test_pred = model.predict(x_test)
    print(" For ",i," Negn Accurcy Score"  , accuracy_score(y_test,y_test_pred))
```

## For Normalized Data

```python
x = minmax.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=12,stratify=y)
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
y_train_pred = model.predict(x_train)
print("Accurcy Score"  , accuracy_score(y_train,y_train_pred))
```

```python
for i in range (1,41,2):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    y_train_pred = model.predict(x_train)
    print(" For ",i," Negn Accurcy Score"  , accuracy_score(y_train,y_train_pred))
```
